# Step 2

```python
import numpy as np
import matplotlib.pyplot as plt

# Time array
x = np.linspace(0, 4*np.pi, 1000)

y1 = np.sin(x)  # Basic sine

# x0 Newton-Raphson setup for x1
x0 = 2.0  # Initial guess (green dot)
f_x0 = np.sin(x0)  # f(x) = sin(x)
f_prime_x0 = np.cos(x0)  # f'(x) = cos(x)

# Tangent line equation: y = f'(x0)(x - x0) + f(x0)
# Extend the tangent line beyond the plot boundaries
tangent_x = np.linspace(-2, 4*np.pi + 2, 200)  # Extended range
tangent_y = f_prime_x0 * (tangent_x - x0) + f_x0
x1 = x0 - f_x0 / f_prime_x0 

# Plot all
plt.figure(figsize=(12, 6))
plt.plot(x, y1, label='sin(x)')
plt.plot(tangent_x, tangent_y, '--', color='orange', label=f'Tangent at x={x1}', alpha=0.7)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='black', linewidth=0.5)
plt.text(3.125, 0.05, 'a', fontsize=10, color='red')
plt.scatter(3.125, 0.0, color='red')
plt.text(2, 0.05, 'b', fontsize=10, color='green')
plt.scatter(2, 0.0, color='green')

plt.text(x1, 0.05, 'c', fontsize=10, color='blue')
plt.scatter(x1, 0.0, color='blue')  # New approximation point
plt.xlim(0, 4*np.pi)
plt.ylim(-1.5, 1.5)

plt.title('Sin Function')
plt.legend()
plt.show()
```

# Step 3

```python
import numpy as np
import matplotlib.pyplot as plt

# Time array
x = np.linspace(0, 4*np.pi, 1000)

y1 = np.sin(x)  # Basic sine

# x0 Newton-Raphson setup for x1
x0 = 2.0
f_x0 = np.sin(x0)
f_prime_x0 = np.cos(x0)
tangent_x = np.linspace(-2, 4*np.pi + 2, 200)
tangent_y = f_prime_x0 * (tangent_x - x0) + f_x0

# x1 Newton-Raphson iteration for x2
x1 = x0 - f_x0 / f_prime_x0  # x-intercept of tangent line
f_x1 = np.sin(x1)  # f(x) = sin(x)
f_prime_x1 = np.cos(x1)  # f'(x) = cos(x)
tangent_x = np.linspace(-2, 4*np.pi + 2, 200)
tangent_y = f_prime_x1 * (tangent_x - x1) + f_x1
x2 = x1 - f_x1 / f_prime_x1  # x-intercept of tangent line

# Plot all
plt.figure(figsize=(12, 6))
plt.plot(x, y1, label='sin(x)')
plt.plot(tangent_x, tangent_y, '--', color='orange', label=f'Tangent at x={x2}', alpha=0.7)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='black', linewidth=0.5)
plt.text(3.125, 0.05, 'a', fontsize=10, color='red')
plt.scatter(3.125, 0.0, color='red')
plt.text(2, 0.05, 'b', fontsize=10, color='green')
plt.scatter(2, 0.0, color='green')
plt.text(x1, 0.05, 'c', fontsize=10, color='blue')
plt.scatter(x1, 0.0, color='blue')  # Old approximation point

plt.text(x2, 0.05, 'd', fontsize=10, color='purple')
plt.scatter(x2, 0.0, color='purple')  # New approximation point
plt.xlim(0, 4*np.pi)
plt.ylim(-1.5, 1.5)

plt.title('Sin Function')
plt.legend()
plt.show()
```
